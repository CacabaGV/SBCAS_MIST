```python
import pandas as pd
import numpy as np
import os

features = ['fc_adm_final', 'plaquetas_adm_final', 'creatinina_adm_final', 'pcr_adm_final', 'sat_fio2']
filepath = '/Users/izadoraganem/Imputation/Imputation/Revista médica/Data/Base_onda_Onda 1_desfecho_obito_variaveis_all.parquet'
output_dir = '/Users/izadoraganem/Imputation/Imputation/Revista médica/MCAR50'

def mcar_50(filepath, features, output_dir, base_seed=7):
    df = pd.read_parquet(filepath)

    for i, feature in enumerate(features):
        seed = base_seed + i
        np.random.seed(seed)

        # copia a base original
        df_copy = df.copy()

        # remove linhas com NaN original na feature
        df_copy = df_copy[df_copy[feature].notna()]

        # número de valores a tornar NaN (50%)
        n_remove = int(len(df_copy) * 0.5)

        # escolhe aleatoriamente índices para aplicar NaN
        idx_missing = np.random.choice(df_copy.index, size=n_remove, replace=False)

        # aplica NaN apenas nos índices selecionados
        df_copy.loc[idx_missing, feature] = np.nan

        # salva o DataFrame com apenas NaNs artificiais
        output_path = os.path.join(output_dir, f"base_mcar50_{feature}.parquet")
        df_copy.to_parquet(output_path, index=True)
        print(f"Gerado MCAR 50% artificial para {feature} (seed={seed}) e salvo em: {output_path}")
        print(f"Total linhas: {len(df_copy)}, NaNs aplicados: {df_copy[feature].isna().sum()}")


mcar_50(filepath, features, output_dir)
```
